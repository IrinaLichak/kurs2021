###Исходныe данныe

```python
#Исходные данные
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

###Блок с функциями

```python
#Перемещение документа с полки на полку
def moving_doc():
  """Перемещаем документ с полки на полку.
  Если номер и полка существуют, 
  то документ удаляется со старой полки и прибавляется на новой.
  Если номер документа или полки не найден, выведется соответствующая надпись.
  """
  nomber_new = input('Введите номер документа: ')
  if nomber_check_directories(nomber_new):
    shelf_new = input('Введите номер полки: ')
    if shelf_new in directories.keys():
      directories[shelf_new].append(nomber_new)
      print('Документ перемещен.')
      print('Текущий список документов:')
      all_info()
    else:
      print('Такой полки не существует. Текущий перечень полок: ', all_shelfs())
  else:
    print('Документ не найден в базе')
    print('Текущий список документов:')
    all_info()
 
#Удаление элемента из documents
def nomber_check_documents(nomber_new):
  """Удаление элемента из documents.
  Если введенный номер есть в значениях, 
  то удалить документ (элемент) из documents.
  """
  i = 0
  for doc in documents:
    for nomber_el in doc.values():
      if nomber_el == nomber_new:
        del(documents[i])
        return True
    i +=1

#Удаление документа с полки
def nomber_check_directories(nomber_new):
  """Удаление документа с полки.
  Если введенный номер есть в значениях, 
  то удалить номер из списка документов на полке.
  """
  for shelf, list_nomber in directories.items():
    if nomber_new in list_nomber:
      list_nomber.remove(nomber_new)
      return True

#Удалить документ по номеру
def del_doc():
  """Удаление документа по номеру.
  Если номер документа есть в значениях documents, то удаляем документ.
  Если номера документа нет, выводит фразу «Документ не найден в базе».
  """
  nomber_new = input('Введите номер документа: ')
  if nomber_check_documents(nomber_new):
    if nomber_check_directories(nomber_new):
      print('Документ удален.')
      print('Текущий список документов:')
      all_info()
  else:
    print('Документ не найден в базе.')
    print('Текущий список документов:')
    all_info()

#Добавить новый документ
"""Ввод в другом порядке (первую вводим полку), 
т.к. не имеет смысла вводить остальную инфу, если полка не существует. 
Так гораздо гуманнее по отношению к пользователю)))
"""
def add_new_doc():
  """Добавляет новый документ.
  Если введенная полка существует, то добавляет новый документ.
  Если полки не существует, то предлагает добавить её через команду as.
  """
  shelf_new = input('Введите номер полки: ')
  if shelf_new in directories.keys():
    new_doc = {}
    new_doc['number'] = input('Введите номер документа: ')
    new_doc['type'] = input('Введите тип документа: ')
    new_doc['name'] = input('Введите владельца документа: ')
    documents.append(new_doc)
    directories[shelf_new].append(new_doc['number'])
    print('Документ добавлен. Текущий список документов:')
    all_info()
  else:
    print('Такой полки не существует. Добавьте полку командой as.')
    print('Текущий список документов:')
    all_info()

#Удалить полку
def del_shelf():
  """Удаляет полки, если такой существует и она пуста.
  Если полки с таким номером нет, выводит фразу «Такой полки не существует».
  Если на полке есть документы, выводит фразу «На полке есть документы, удалите их перед удалением полки».
  """
  shelf_new = input('Введите номер полки: ')
  if shelf_new not in directories.keys():
    return print('Такой полки не существует. Текущий перечень полок: ', all_shelfs()) 
  else:
    if directories[shelf_new] == []:
      del(directories[shelf_new])
      return print('Полка удалена. Текущий перечень полок:', all_shelfs())
    else:
      return print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:', all_shelfs())

#Добавление полки
def add_shelf():
  """Добавляет полку, если такой не существует.
  Иначе выводит фразу «Такая полка уже существует».
  """
  shelf_new = input('Введите номер полки: ')
  if shelf_new not in directories.keys():
    directories[shelf_new] = []
    return print('Полка добавлена. Текущий перечень полок: ', all_shelfs())
  else:
    return print('Такая полка уже существует. Текущий перечень полок:', all_shelfs())

#Перечислить все полки
def all_shelfs():
  """Выводит перечень имеющихся полок.
  Формирует список из ключей (номеров полок) directories.
  Обращает в строку и выводит на экран.
  """
  shelf_all = directories.keys()
  shelf_all = ', '.join(shelf_all)
  return shelf_all

#Вывести всю информацию о всех документах
def all_info():
  """Выводит всю информацию по всем имеющимся документам.
  """
  dict_info = {}
  for dict_el in documents:
    dict_info['№'] = dict_el['number']
    dict_info['тип'] = dict_el['type']
    dict_info['владелец'] = dict_el['name']
    for shelf, list_el in directories.items():
      if dict_el['number'] in list_el:
        dict_info['полка хранения'] = shelf
    list_info = []
    for key_el, values_el in dict_info.items():
      list_info.append("{}: {}".format(key_el, values_el))
    info_all_doc = ", ".join(list_info)
    print(info_all_doc)
    dict_info = {}

#Узнать имя владельца документа по номеру
def name_by_number():
  """Поиск владельца по номеру документа.
  Если номер документа есть в значениях documents, то выводит имя владельца этого документа.
  Если номера документа нет, выводит фразу «Документ не найден в базе».
  """
  nomber_new = input('Введите номер документа: ')
  for doc in documents:
    for nomber_el in doc.values():
      if nomber_el == nomber_new:
        return print('Владелец документа:', doc['name'])
  return print('Документ не найден в базе')

#Узнать полку хранения по номеру документа
def shelf_by_number():
  """Выводит номер полки по номеру документа.
  Если номер документа есть среди значений directories, то выводит номер полки.
  Если номера документа нет, выводит фразу «Документ не найден в базе».
  """
  nomber_new = input('Введите номер документа: ')
  for shelf, list_nomber in directories.items():
    if nomber_new in list_nomber:
      return print('Документ хранится на полке:', shelf)
  return print('Документ не найден в базе')
```

```python
#Цикличный ввод. Проверка команд.
def main():
  """Цикличный ввод и запуск команд.
  Проверяет введенную команду и запускает соответствующее действие.
  Прекратить ввод команд можно командой 'q'.

  Перечень доступных команд:
  'q': 'Закончить ввод команд',
  'p': 'Поиск владельца по номеру документа',
  's': 'Поиск полки по номеру документа',
  'l': 'Вывести всю информацию',
  'as': 'Добавить полку',
  'ds': 'Удалить полку',
  'ad': 'Добавить документ',
  'd': 'Удалить документ'
  """
  while True:
    comand = input('Введите команду: ')
    if comand == 'q':
      break
    elif comand == 'p':
      name_by_number()
    elif comand == 's':
      shelf_by_number()
    elif comand == 'l':
     all_info()
    elif comand == 'as':
      add_shelf()
    elif comand == 'ds':
      del_shelf()
    elif comand == 'ad':
      add_new_doc()
    elif comand == 'd':
      del_doc()
    elif comand == 'm':
      moving_doc()
    else:
      print('Неверная команда')
```

###Запуск программы

```python
main()
```
